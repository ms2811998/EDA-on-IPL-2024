```python
#sample code
import pandas as pd

# Assuming df is your pandas DataFrame containing match data

# Create a dictionary to store winning streaks for each team
winning_streaks = {team: 0 for team in ipl_df['First_Team'].unique()}

# Iterate through each team
for team in winning_streaks.keys():
    # Filter matches where the team played as either the first team or the second team
    team_matches = ipl_df[(ipl_df['First_Team'] == team) | (ipl_df['Second_Team'] == team)]

   
    # Initialize variables to track winning streak
    current_streak = 0
    max_streak = 0
    
    # Iterate through the matches involving the team
    for index, match in team_matches.iterrows():
        # Check if the team won the match
        if match['Winning_team'] == team:
            current_streak += 1
            # Update maximum streak if the current streak is greater
            max_streak = max(max_streak, current_streak)
        else:
            # Reset streak if the team lost the match
            current_streak = 0
    
    # Update the winning streak for the team in the dictionary
    winning_streaks[team] = max_streak

# Print the winning streaks for each team
for team, streak in winning_streaks.items():
    print(f"{team}: {streak}")
```

```python
#First Inning Highest Score Team Wise
import re
dict1={team:0 for team in ipl_df.First_Team.unique()}

for team in dict1.keys():
    first_inning_matches=ipl_df[ipl_df.First_Team==team]
    current_score=0

    for index,row in first_inning_matches.iterrows():
        first_inning_score=row['First Innings Score']
        match=re.search(r'(\d+)', first_inning_score)
        if match:
            first_inning_score=int(match.group())
            if first_inning_score>current_score:
                current_score=first_inning_score
        #else:
            #current_score=current_score
    dict1[team]=current_score
print(dict1)
        
```

```python
#Second Innings Highest Score Team Wise
import re
dict2={team:0 for team in ipl_df.Second_Team.unique()}

for team in dict2.keys():
    second_inning_matches=ipl_df[ipl_df.Second_Team==team]
    current_score=0
    
    for index,row in second_inning_matches.iterrows():
        second_inning_score=row['Second Innings Score']
        match=re.search(r'(\d+)',second_inning_score)
        if match:
            second_inning_score=int(match.group())
            if second_inning_score > current_score:
                current_score=second_inning_score
    
    dict2[team]=current_score
print(dict2)
```

```python
# Combinedly in both the innings, for each team, what is the highest score
dict3={}
for team in dict1.keys():
    if dict1[team]>dict2[team]:
        dict3[team]=dict1[team]
    else:
        dict3[team]=dict2[team]
print(dict3)
```

```python
print('Q7. How many times does a team win after crossing 200 score?')
```

```python
dict={team:0 for team in ipl_df.First_Team.unique()}

for team in dict.keys():
    team_matches=ipl_df[(ipl_df.First_Team==team) | (ipl_df.Second_Team==team)]
    count1,count2=0,0
    for index,row in team_matches.iterrows():
        score1=row['First Innings Score']
        score2=row['Second Innings Score']
        match1=re.search(r'(\d+)',score1)
        match2=re.search(r'(\d+)',score2)
        if match1:
            score1=int(match1.group())
            if score1>200:
                count1+=1
                if team==row['Winning_team']:
                    count2+=1 
        if match2:
            score2=int(match2.group())
            if score2>200:
                count1+=1
                if team==row['Winning_team']:
                    count2+=1 
        
    dict[team]=[count1,count2]
print(dict)
            
```
